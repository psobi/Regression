```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle5 as pickle
```

```python
file_train_X = open("../data/X_train_03_preprocess","rb")
file_train_y = open("../data/y_train_03_preprocess","rb")

X = pickle.load(file_train_X)
y = pickle.load(file_train_y)

X.isnull().sum().sum() #No missing values
X.shape
```

```python
X.columns
```

## Random Forest for Relative Feature Importance
Parameters: tree estimators 100, max_depth=8, criterion='gini'

```python
from sklearn.ensemble import RandomForestClassifier

model_feature_selection = RandomForestClassifier(n_estimators=500,criterion='gini',max_depth=15)
m = model_feature_selection.fit(X,y)
```

## Cumulative Feature Relative Importance

```python
feats = m.feature_importances_
cols = X.columns.values

feature_importance = pd.DataFrame(data=feats, columns=['Feature Importance'])
feature_name = pd.DataFrame(data=cols, columns=['Feature Name'])
frames = [feature_importance, feature_name]

f = pd.concat(frames, axis=1, sort=False)
fi_cum = f.sort_values(by=['Feature Importance'],ascending=False)

fi_cum.to_excel("output_feature_importance.xlsx")

fi_cum['Cumulative Feature Importance'] = fi_cum['Feature Importance'].cumsum()
fi_cum['Cumulative Feature Importance'] = 100*fi_cum['Cumulative Feature Importance']/fi_cum['Feature Importance'].sum()

pd.set_option('display.max_rows', 100)

feature_importance = fi_cum
feature_importance
```

35 features (50%) account for 95% variability in target variable
25 features (30%) account for 90% variability in target variable
55 features account for 99.1% 
totoal 79 features

```python
feature_importance_threshold = 99.0

features_below_treshold = feature_importance[feature_importance['Cumulative Feature Importance'] <= feature_importance_threshold]
features_above_treshold = feature_importance[feature_importance['Cumulative Feature Importance'] > feature_importance_threshold]
```

```python
import pylab as plot
params = {'legend.fontsize': 15,
          'legend.handlelength': 2}
plt.rcParams['xtick.labelsize']=12
plt.rcParams['ytick.labelsize']=14

plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['legend.fontsize'] = 14

plt.rcParams['axes.labelsize'] = 14
plt.rcParams['axes.labelweight'] = 'bold'

plot.rcParams.update(params)

sns.set(rc={'figure.figsize':(14,8)})


import matplotlib.pylab as plt
params = {'legend.fontsize': 14,
          'figure.figsize': (14, 7),
         'axes.labelsize': 14,
         'axes.titlesize':14,
         'xtick.labelsize':12,
         'ytick.labelsize':14}
plt.rcParams.update(params)

data = feature_importance.sort_values(by=['Feature Importance'],ascending=False)
ax = sns.barplot(x='Feature Name',y='Feature Importance', data=data[:100],palette='RdYlGn_r')

ax.set_title('Feature Relative Importance',fontsize=16)
plt.xlabel('Feature Name')
plt.ylabel('Feature Relative Importance')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

ax2 = ax.twinx()
data.plot(x="Feature Name", y="Cumulative Feature Importance", ax=ax2, legend=True, color="r", linewidth = 3)

#ax2.axhline(feature_importance_threshold, linestyle ="--", color="black")
ax2.axvline(len(features_below_treshold), linestyle ="--", color="black")
ax2.set_ylabel('Cumulative Importance')
ax2.tick_params(axis='y', labelsize=14)
#ax2.tick_params(axis='y', size=16)


plt.legend(['75% features account for\n 99% variablity in Target'],loc='center')
plt.tight_layout()
plt.show()
figure = ax.get_figure()    

figure.savefig('../doc/figures/feature_importance.png', dpi=600)
```
