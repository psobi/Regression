```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline

#import warnings
import pickle5 as pickle
#warnings.filterwarnings('ignore')
```

```python
#from sklearn.model_selection import train_test_split, GridSearchCV
#from sklearn.metrics import confusion_matrix,accuracy_score, roc_curve, auc
sns.set_style("whitegrid")
#Import data
import pickle5 as pickle

file_train_X = open("../data/fe_train_X","rb")
file_train_y = open("../data/fe_train_y","rb")

X_train = pickle.load(file_train_X)
y_train = pickle.load(file_train_y)

X_train.isnull().sum().sum() #No missing values
X_train.shape
```

```python
X_ = X_train.copy()

X_["TotalBsmtFin"] = X_["BsmtFinSF1"] + X_["BsmtFinSF2"]
X_["TotalBath"] = X_["FullBath"] + 0.5*X_["HalfBath"] + X_["BsmtFullBath"] + 0.5*X_["BsmtHalfBath"]

#X_["TotalLivArea"] = X_["GrLivArea"] + X_["TotalBsmtFin"]
    
drop_columns = ['BsmtFinSF1','BsmtFinSF2','FullBath','HalfBath','BsmtFullBath','BsmtHalfBath', '1stFlrSF','2ndFlrSF','1stFlrSF','2ndFlrSF', 'GarageCars','GarageYrBlt']


X_fs = X_.drop(drop_columns, axis=1)
X_fs.columns
```

```python
from sklearn.ensemble import RandomForestClassifier
model_feature_selection = RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=8)
m = model_feature_selection.fit(X_fs,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_test = Pipeline(steps=[('classifier', RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=8))])

m = model_test.fit(X_fs,y_train)
feats = m.named_steps["classifier"].feature_importances_
cols = X_fs.columns.values
len(cols),len(feats)

feature_importance = pd.DataFrame(data=feats, columns=['Feature Importance'])
feature_name = pd.DataFrame(data=cols, columns=['Feature Name'])

frames = [feature_importance, feature_name]

f = pd.concat(frames, axis=1, sort=False)
fi = f.sort_values(by=['Feature Importance'],ascending=False)

##d['Feature Importance'].sum()
fi.to_excel("output_feature_importance_4.xlsx")
#1stFlrSF
fi['Cumulative Feature Importance'] = fi['Feature Importance'].cumsum()
fi['Cumulative Feature Importance Pct'] = 100*fi['Cumulative Feature Importance']/fi['Feature Importance'].sum()
pd.options.display.max_rows
pd.set_option('display.max_rows', 100)
fi
```

```python
fi_subset = fi[fi['Cumulative Feature Importance Pct'] <= 99.1]
fi_subset

cols_list = fi_subset['Feature Name'].tolist()

X_train_selected_features_for_model = X_[cols_list]
```

```python
fi_subset = fi[fi['cum_perc'] <= 90.0]
fi_subset

cols_list = fi_subset['Feature Name'].tolist()

X_train_selected_features = X_[cols_list]
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import DivergingNorm
import numpy as np

plt.figure(figsize=(6,6))

sns.set(style='whitegrid', color_codes=True)

data = fi_subset.sort_values(by=['Feature Importance'],ascending=False)

#data['Feature Importance'].min()
data['Feature Importance'].max()
#0.003664472900658729

#data
#norm = DivergingNorm(vmin=data['Feature Importance'].min(), vcenter=0, vmax=data['Feature Importance'].max())
norm = DivergingNorm(vmin=0.003664472900658729, vcenter=0.06, vmax=0.1274039927746411)

colors = [plt.cm.RdYlGn(norm(c)) for c in data['Feature Importance']]
ax = sns.barplot(x='Feature Importance', y='Feature Name', data=data, palette=colors, edgecolor='black')

ax.set_title('Feature Relative Importance', size = 16,fontweight='bold')
ax.set_ylabel('Feature Name',size = 14,fontweight='bold')
ax.set_xlabel('Feature Importance',size = 14,fontweight='bold')
ax.tick_params(labelsize=12)
#plt.legend(['1 - High Value\n0 - Low Value'],loc='upper left')

plt.tight_layout()
plt.show()

figure = ax.get_figure()    
figure.savefig('../doc/figures/feature_importance.png', dpi=600)
```
